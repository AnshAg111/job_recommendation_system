```python
def get_mean_experience(exp):
    first_number_index =exp.find(' ')
    first_number = int(exp[:first_number_index].strip())
    second_number_index = exp.find(' ',first_number_index+1)
    stop_index= exp.find('Years')
    second_number = int(exp[second_number_index:stop_index].strip())
    mean_exp = (first_number+second_number)/2
    return mean_exp

df['mean_experience'] = df['Experience'].map(get_mean_experience)
df.head()
```

```python
def get_mean_salary(sal):
    sal = sal.strip()
    first_number_index = sal.find('$')+1
    first_number_stop = sal.find('K')
    first_number = int(sal[first_number_index:first_number_stop])
    second_number_index = sal.find('$',first_number_index+1)+1
    second_number_stop = sal.find('K',first_number_stop+1)
    second_number = int(sal[second_number_index:second_number_stop])
    mean_salary = (first_number+second_number)/2
    return mean_salary
```

```python
df['mean_salary'] = df['Salary Range'].map(get_mean_salary)
df.head()
```

```python
df.drop(columns=['Job Id','Contact Person','Contact'],inplace=True)
```

```python
df['Salary Range'] = df['Salary Range'].str.replace('$', '')
df['Salary Range'] = df['Salary Range'].str.replace('K', '000')
df['Salary Range'] = df['Salary Range'].str.replace('-', ' ').str.split(' ')
df['Salary Range'] = df['Salary Range'].apply(lambda x: [int(i) if i.isdigit() else pd.NA for i in x])
df['Min Salary'] = df['Salary Range'].apply(lambda x: x[0] if x and len(x) >= 1 else pd.NA)
df['Max Salary'] = df['Salary Range'].apply(lambda x: x[1] if x and len(x) >= 2 else pd.NA)

df.drop(columns=['Salary Range'], inplace=True)
df.head()
```
